# Bands analysis
This script imports the band values (divided by class) from the Earth Engine classification and plots them on a KDE graph.

```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

Navigate to the folder where the data (in csv format) is contained and import the data.

```python
mypath = './data/class_bands/'
files = [f.split('.')[0] for f in listdir(mypath) if isfile(join(mypath, f))]

dataframe={}
for file in files:
     dataframe[file] = pd.read_csv(mypath+file+'.csv')#provide separator if required

print ('The available classes are:')
for key, value in dataframe.items() :
    print (key)
```

```python
plt.plot(dataframe['vegetation']['Band Value'],dataframe['vegetation']['Red'])
plt.plot(dataframe['vegetation']['Band Value'],dataframe['vegetation']['Blue'])
plt.plot(dataframe['vegetation']['Band Value'],dataframe['vegetation']['Green'])
plt.plot(dataframe['vegetation']['Band Value'],dataframe['vegetation']['NIR'])
```

```python
plt.plot(dataframe['built_environment']['Band Value'],dataframe['built_environment']['Red'])
plt.plot(dataframe['built_environment']['Band Value'],dataframe['built_environment']['Blue'])
plt.plot(dataframe['built_environment']['Band Value'],dataframe['built_environment']['Green'])
plt.plot(dataframe['built_environment']['Band Value'],dataframe['built_environment']['NIR'])
```
